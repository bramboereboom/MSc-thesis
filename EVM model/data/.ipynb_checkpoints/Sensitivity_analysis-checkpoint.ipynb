```python

df1 = pd.read_csv('sensitivity_run1_evs0.9_price0.9.csv')
df2 = pd.read_csv('sensitivity_run2_evs0.9_price1.csv')
df3 = pd.read_csv('sensitivity_run3_evs0.9_price1.1.csv')
df4 = pd.read_csv('sensitivity_run4_evs1_price0.9.csv')
df5 = pd.read_csv('sensitivity_run5_evs1_price1.csv')
df6 = pd.read_csv('sensitivity_run6_evs1_price1.1.csv')
df7 = pd.read_csv('sensitivity_run7_evs1.1_price0.9.csv')
df8 = pd.read_csv('sensitivity_run8_evs1.1_price1.csv')
df9 = pd.read_csv('sensitivity_run9_evs1.1_price1.1.csv')

df10 = pd.read_csv('sensitivity_run1_evs0.8_price0.8.csv')
df11 = pd.read_csv('sensitivity_run2_evs0.8_price1.2.csv')
df12 = pd.read_csv('sensitivity_run3_evs1.2_price0.8.csv')
df13 = pd.read_csv('sensitivity_run4_evs1.2_price1.2.csv')



df_list = [df1,df2,df3,df4,df5,df6,df7,df8,df9,df10,df11,df12,df13]
```

```python
for i in df_list:
    print(i.total_current_power_demand.mean())
```

```python
def plot_sensitivity(KPI):
    x = np.linspace(0,60,60)
    for i in df_list:
        plt.plot(x,i[KPI][240:300])
    
    plt.title('EVM sensistivity analysis plot')
    plt.ylabel('VTG capacity (KW)')
    plt.xlabel('Time (15 min)')
    
    plt.show()
```

```python
plot_sensitivity('total_VTG_capacity')
```
