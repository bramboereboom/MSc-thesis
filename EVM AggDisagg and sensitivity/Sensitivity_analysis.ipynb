```python
import pandas as pd
import matplotlib.pyplot as plt
import itertools
import numpy as np
import statistics as stats
```

```python
def data_extraction(name,scenarios, repetitions,KPI):
    for h in range(len(KPI)):
        for i in range(len(scenarios)):
            Means = []
            for j in range(repetitions):
                df = pd.read_csv('{}_scen{}_rep{}_evs{}_price1.csv'.format(name,i+1,j+1,scenarios[i]))
                
                Means.append(df[KPI[h]][160:450].mean())
            print('EVs {},KPI {}, mean {}'.format(scenarios[i],KPI[h],stats.mean(Means)))
            
    
```

```python
data_extraction('Price_disagg_12',[0.9,1,1.1],1,['total_current_power_demand','total_VTG_capacity'])
#'total_VTG_capacity'
#'total_current_power_demand'
```
