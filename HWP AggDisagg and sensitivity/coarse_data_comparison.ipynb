# This notebook contains the functions using in the coarsening sensitivity analysis of the HWP

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# load HWP run results with original and coarsed input data

df_normal = pd.read_csv('normal_base_case.csv')
df_coarse = pd.read_csv('coarse_base_case.csv')

df_normal = df_normal.drop([0,1,2,3,4])
df_coarse = df_coarse.drop([0,1,2,3,4])
```

```python
# plot behaviour difference of cashflow KPI per hour

plt.plot(df_normal['T'][:240], df_normal['cashflow WA'][:240], label = 'Base case data')
plt.plot(df_coarse['T'][:240], df_coarse['cashflow WA'][:240], label = 'Data with time A/D' )
plt.xlabel('Time (hours)')
plt.ylabel('Cashflow (€)')
plt.title('Cashflow comparison')
plt.legend()
```

```python
# plot behaviour difference of hydrogen use

plt.plot(df_normal['T'][:120], df_normal['gebruik waterstof'][:120], label = 'Base case data')
plt.plot(df_coarse['T'][:120], df_coarse['gebruik waterstof'][:120], label = 'Data with time A/D' )
plt.title('Hydrogen use comparison')
plt.xlabel('Time (hours)')
plt.ylabel('Cashflow (€)')
plt.legend(loc='upper left')
plt.rcParams["figure.figsize"] = (20,10)
```
